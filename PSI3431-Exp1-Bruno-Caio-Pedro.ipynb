# Item a) Amostragem do Sinal



<span style="font-family: 'Times New Romans'; font-size: 1.4em">

O sinal de tempo contínuo $x_{0}(t) = (sin(Ω * t))^3*e^{(-t/τ)}$ foi amostrado por uma frequência de amostragem:


$f_a = 40 kHz$, implicando em $T_a = \frac{1}{f} = 0,0000025 s$, 

Assim, foi obtido o sinal de tempo discreto $x_{0}(t) = (sin(Ω * \frac{n}{fa}))^3*e^{(-\frac{n}{fa}/τ)}$

Logo, no intervalo de tempo de 0 a 2 segundos temos que há 80.000 amostras e, portanto, o comprimento de $x_{0}[n]$ será N = $80.001$

```julia
#fa = 40kHz -> Ta = 0,0000025 s -> Há 80k amostras entre 0 e 2s
N = 80001
n = 0:N-1
Ω = 2*pi*500
fa = 40e3
τ = 0.5
x0 = ((sin.(Ω * n / fa)).^3).*exp.(-(n/fa)/τ)
plot(n,x0)
grid()
xlabel("n")
ylabel(L"$x_0$")
title(L"Função $x_0[n]$");
wavplay(x0, fa)
```

# Item b) Adição do ruído



<span style="font-family: 'Times New Romans'; font-size: 1.4em">


Sabendo que potência de um sinal pode ser representado por $c_x = E\{x^2\} - E\{x\}^2$  

Temos que $c_x = r_x - \mu_x$ e , por se tratar de uma função senoide, que $\mu_x = 0$.

Logo, podemos calcular a potência do sinal $x_0[n]$ elevando todos os elementos ao quadrado e calculando a média, segundo a fórmula: 

Potência média de $x_{0} = Pot_{x_{0}}= E \left \{ x_{0}[n]^2 \right \}$ 

Para um SNR igual a 10 dB, precisamos que a potência de $x_0$ seja 10 vezes maior que a de $v_x$ pois  
$SNR = 10 \cdot log_{10}(\frac{Pot_{x_{0}}}{Pot_{v_{x}}})$  
$10 = 10 \cdot log_{10}(\frac{Pot_{x_{0}}}{Pot_{v_{x}}})$  
$Pot_{x_{0}} = 10 Pot_{v_{x}}$  

Sabemos também que se a variância $\sigma ^2$ de um sinal é igual à sua potência nos casos onde o valor esperado é zero, então podemos representar seu desvio padrão $\sigma_{v_{x}}$ como $\sigma_{v_{x}} = \sqrt{Pot_{v_{x}}}$

Para fazer um ruído gaussiano com desvio padrão diferente de 1, multiplicamos $\sqrt{Pot_v}$ por todos os termos do vetor gerado pela função randn, cuja distribuição é gaussiana normalizada com desvio padrão 1 e valor esperado 0, pois $\sigma_{1,2} = \sigma_1 \cdot \sigma_2$

```julia
pot_x = mean(x0.^2)
pot_v = pot_x/10 # SNR igual a 10 dB
v = sqrt(pot_v).*randn(N,1); #sigma_v = sqrt(pot_v)
wavplay(v, fa)
```

```julia
x = x0+v
plot(n, x, label = "x[n]")
grid()
xlabel("n")
ylabel(L"$x[n]$ e $x_0[n]$")
title(L"Função $x[n]$ sobreposta com $x_0[n]$");
plot(n, x0, label = L"$x_0[n]$")
legend()
wavplay(x, fa) #ouvindo o sinal x[n]
```

# Item e) Classificação quanto à característica estacionária



<span style="font-family: 'Times New Romans'; font-size: 1.4em">

Os critérios para se definir um processo como estacionário no sentido amplo (WSS) são: 
1. $\mu_{x}[n]$ ser constante
2. $r_{x}[n,n+l]$ não variar em função de n, ou seja, $r_{x}[n,n+l] = r_{x}[l]$
    
Dessa forma, x[n] e y[n] podem ser excluídos pois nenhum dos critérios é atendido, já que tanto a Função Valor Esperado quanto a Função Autocorrelação variam em função de n. 

Isso ocorre pois os sinais podem ser decompostos por: $x[n] = x_{0}[n] + v_{x}[n]$ e $y[n] = y_{0}[n] + v_{y}[n]$, sendo $x_{0}[n]$ e $y_{0}[n]$ componentes determinísticas cujo valor esperado varia de acordo com n.

Além disso, como todo processo estacionário no sentido estrito (SSS) é também WSS, pode-se afirmar que x[n] e y[n] não são estacionários em nenhum sentido.
    
No caso dos ruídos, as subtrações $x[n]-x_{0}[n]$ e $y[n]-y_{0}[n]$ cancelam as parcelas determinísticas dos processos anteriores, fazendo com que eles se encaixem nos critérios WSS, ou seja, possuem valor esperado e variância que não variam com o deslocamento no tempo ou na amostragem, visto que ambos os ruídos possuem distribuição gaussiana com valor esperado nulo.

Assim, os ruídos são processos estacionários WSS enquanto que os sinais $x[n]$ e $y[n]$ não são estacionários em nenhum sentido devido à parcela determinística

# Item f) Densidade Espectral de Potência e Potência



<span style="font-family: 'Times New Romans'; font-size: 1.4em">

Densidade espectral de potência do ruído na entrada: 
    
\begin{equation*}
S_{v_{x}} ( \omega )=  \sum_{l = -\infty}^{\infty}r_{v_{x}}[l] \cdot e^{-j\omega l}
\end{equation*}

\begin{equation*}
r_{v_{x}}[l] = E\{V[n+l] \cdot V[n]\} = 
\begin{cases}
0 \quad se \quad l \neq 0 \\
\sigma ^{2} \quad se \quad l = 0 
\end{cases}
\end{equation*}
    
\begin{equation*}
S_{v_{x}} ( \omega ) =  \sigma _{v_{x}} ^{2} 
\end{equation*}

Densidade espectral de potência do ruído na saída:

\begin{equation*}
S_{v_{y}} ( \omega ) =  S_{v_{x}} ( \omega ) \cdot |H(e^{j \omega })|^{2}
\end{equation*}
    
Potência do ruído na entrada:

\begin{equation*}
Pot_{média} = \frac{1}{2 \pi } \int_{- \pi }^{\pi } S_{v_{x}} ( \omega )  d \omega = \sigma ^{2} _{v_{x}}
\end{equation*}
    
Potência do ruído na saída:

\begin{equation*}
Pot_{média} = \sigma ^{2} _{v_{y}} 
    = \frac{1}{2 \pi } \int_{- \pi }^{\pi } S_{v_{x} ( \omega )} |H(e^{j \omega })|^{2}  d \omega 
\end{equation*}
    
Como o ruído é branco:
    
\begin{equation*}
Pot_{média} 
    = \frac{1}{2 \pi } \sigma ^{2} _{v_{x}} \int_{- \pi }^{\pi }  |H(e^{j \omega })|^{2}  d \omega 
\end{equation*}
    
Aplicando o Teorema de Parseval:
    
\begin{equation*}
Pot_{média} 
    = \sigma ^{2} _{v_{x}} \cdot \sum^{\infty}_{n=-\infty} h^{2}[n]
\end{equation*}
    
Como o filtro é FIR e causal:
    
\begin{equation*}
Pot_{média} 
    = \sigma ^{2} _{v_{x}} \cdot \sum^{N-1}_{n=0} h^{2}[n]
\end{equation*}    

Pelo código abaixo, foram calculados os valores teóricos de potência média na entrada e saída como:

Valor teórico da potência média de $v_{x}$ = $Pot_{v_{x}} = 0.0039048886338914368$

Valor teórico da potência média de $v_{y}$ = $Pot_{v_{y}} = 0.00038259176085767493$
   

```julia
#Potência na entrada
pot_v_x =  pot_v 

#Potência na saída
pot_v_y = pot_v * sum(h.^2)

print("Valor teórico da potência média de vx = ", pot_v_x)
print("\n")
print("Valor teórico da potência média de vy = ", pot_v_y)

```

# Item g) (a) Processo Estocástico


Pelo item e), sabemos que os ruídos $v_x[n]$ e $v_y[n]$ são estacionários no sentido amplo (WSS).

Assim, podemos representar a "autocorrelação amostral" como:

\begin{equation*}
    r_{vx}[l] = E\left \{ V_x[n]*V_x[n+l] \right \}   
\end{equation*}

\begin{equation*}
   r_{vx}[0] = E\left \{ V_x[n]*V_x[n+0] \right \} = E\left \{ (V_x[n])^{2} \right \}
\end{equation*}


Considerando processo independente e igualmente distribuído (iid), temos que a autocorrelação amostral em função de n pode ser aproximada por:

\begin{equation*}
r_{vx}[0] = \sum_{l = 1}^{L}(v_{x}^{l}[n])^{2}\cdot p_{l} = \frac{1}{L} \sum_{l = 1}^{L}(v_{x}^{l}[n])^{2}
\end{equation*}



```julia
L = 1000
N = 80001
n = 0:N-1

r_vx = zeros(N)
r_vy = zeros(N)

v = sqrt(pot_v).*randn(L,N); #sigma_v = sqrt(pot_v), matriz de L linhas e N colunas. Cada linha L é uma observação do processo.
#wavplay(v[1,:], fa)

for l in 1:L
    v_x = v[l, :]
    v_y = filt(h, v_x)
    
    #Autocorrelação amostral (em função de n)
    r_vx = r_vx + v_x.^2
    r_vy = r_vy + v_y.^2
end
```

```julia
#Cálculo da Autocorrelação Amostral (ou seja, a média de diversas realizações dos processos em função de N)
r_vx = r_vx/L
r_vy = r_vy/L

subplot(211)
stem(n, r_vx)

subplot(212)
stem(n, r_vy)

r_vx_media = sum(r_vx)/N

r_vy_media = sum(r_vy)/N

print("r_vx_media = ", r_vx_media)
print("\n")
print("r_vy_media = ", r_vy_media)
```

# Item g) (b)



<span style="font-family: 'Times New Romans'; font-size: 1.4em">
Novamente pelo item e), sabemos que os ruídos $v_x[n]$ e $v_y[n]$ são estacionários no sentido amplo (WSS).

Como o processo é estacionário (WSS) e gaussiano com variância finita, pode-se considerá-lo independente e igualmente distribuído (iid) e, consequentemente, ergódico de segunda ordem, ou seja:

\begin{equation*}
    Pr\left \{ r_{v_{x}}[l] = \bar{r_{v_{x}}}[l] \right \} = 1
\end{equation*}

Assim, podemos representar a autocorrelação média ("autocorrelação temporal") como:

\begin{equation*}
    \bar{r_{v_{x}}}[l] = \lim_{N \rightarrow \infty  }\frac{1}{2N+1}\sum_{n=-N}^{N} v_x[n]*v_x[n+l]
\end{equation*}

Para L = 0 e considerando o intervalo de 0 a N, temos a equação correspondente:

\begin{equation*}
   \bar{r_{v_{x}}}[0] = \lim_{N \rightarrow \infty  }\frac{1}{N}\sum_{n=0}^{N} (v_x[n])^2
\end{equation*}

Essa expressão pode ser calculada para cada realização l, sendo tirada a média entre os valores de $ \bar{r_{v_{x}}}[0] $ para as L observações.

```julia
L = 1000
N = 80001
n = 0:N-1

r_vx_amostral = zeros(N)
r_vy_amostral = zeros(N)
r_vx_temporal = zeros(L) #Cada elemento corresponde ao r_vx de uma realização
r_vy_temporal = zeros(L) #Cada elemento corresponde ao r_vx de uma realização


for l in 1:L
    v = sqrt(pot_v).*randn(N,1); #sigma_v = sqrt(pot_v), matriz de 1 linhas e N colunas. A cada laço do for é uma observação
    v_x = v
    v_y = filt(h, v_x)
    
    r_vx_temporal[l] = sum(v_x.^2)/N
    r_vy_temporal[l] = sum(v_y.^2)/N

    r_vx_amostral = r_vx_amostral + v_x.^2
    r_vy_amostral = r_vy_amostral + v_y.^2
end

#Cálculo da Autocorrelação Amostral (ou seja, a correlação de diversas realizações dos processos em função de N)
r_vx_amostral = r_vx_amostral/L
r_vy_amostral = r_vy_amostral/L

subplot(211)
stem(n, r_vx_amostral)

subplot(212)
stem(n, r_vy_amostral)

r_vx_amostral_media = sum(r_vx_amostral)/N

r_vy_amostral_media = sum(r_vy_amostral)/N

print("\n")
print("r_vx_amostral_media = ", r_vx_amostral_media)
print("\n")
print("r_vy_amostral_media = ", r_vy_amostral_media)
print("\n")

#Cálculo da Autocorrelação Temporal (ou seja, a correlação de cada realização L)

#print(r_vx_temporal)
#print("\n")
#print(r_vy_temporal)

r_vx_temporal_media = sum(r_vx_temporal)/L #Média da autocorrelação de cada realização L de vx

r_vy_temporal_media = sum(r_vy_temporal)/L #Média da autocorrelação de cada realização L de vy

print("\n")
print("r_vx_temporal_media = ", r_vx_temporal_media)
print("\n")
print("r_vy_temporal_media = ", r_vy_temporal_media)
print("\n")

#Cálculo da diferença entre os valores experimentais e teóricos
print("\n")
print("Desvio (%) entre experimental e teórico para vx = ", ((max(r_vx_temporal_media,pot_v_x)-min(r_vx_temporal_media,pot_v_x))/max(r_vx_temporal_media,pot_v_x))*100," %")
print("\n")
print("Desvio (%) entre experimental e teórico para vy = ", ((max(r_vy_temporal_media,pot_v_y)-min(r_vy_temporal_media,pot_v_y))/max(r_vy_temporal_media,pot_v_y))*100," %")
print("\n")
```
